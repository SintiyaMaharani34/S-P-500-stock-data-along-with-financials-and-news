```
import pandas as pd

# =========================
# 1. LOAD DATA
# =========================
df = pd.read_csv('/content/global_ads_performance_dataset.csv')

# lihat data awal
print(df.head())
print(df.info())
```

```
# =========================
# 2. STANDARDIZE COLUMN NAME
# =========================
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

print(df.columns)
```

```
# =========================
# 3. HANDLE MISSING VALUES
# =========================

# cek null
print(df.isnull().sum())

# kolom penting → drop kalau kosong
df = df.dropna(subset=['impressions', 'clicks', 'ad_spend'])

# isi null numerik dengan 0
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].fillna(0)
```

```
# =========================
# 4. FIX DATA TYPE
# =========================
# convert date kalau ada
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
```

```
# =========================
# 5. REMOVE DUPLICATES
# =========================
df = df.drop_duplicates()

```

```
# =========================
# 6. FILTER DATA ANEH
# =========================
df = df[df['impressions'] > 0]
df = df[df['ad_spend'] >= 0]
df = df[df['clicks'] >= 0]

# ctr harus logis
df = df[df['ctr'] <= 1]
```

```
# =========================
# 7. CREATE KPI COLUMNS ⭐
# =========================

df['ctr'] = df['clicks'] / df['impressions']
df['cpc'] = df['ad_spend'] / df['clicks'].replace(0, 1)
df['cpa'] = df['ad_spend'] / df['conversions'].replace(0, 1)
df['roas'] = df['revenue'] / df['ad_spend'].replace(0, 1)
```

```
df['ctr_check'] = df['clicks'] / df['impressions']
df['cpc_check'] = df['ad_spend'] / df['clicks'].replace(0,1)
df['roas_check'] = df['revenue'] / df['ad_spend'].replace(0,1)

print(df[['ctr', 'ctr_check']].head())
```

```
# =========================
# 8. HANDLE OUTLIERS (OPTIONAL)
# =========================
# buang data extreme (optional)
df = df[df['ctr'] <= 1]

```

```
# =========================
# 9. EXPORT CLEAN DATA
# =========================
df.to_csv('ads_clean.csv', index=False)

print("✅ Data sudah bersih & siap analisis!")
```
