3. Extract File ZIP

```
import zipfile

with zipfile.ZipFile('/content/price_data.zip', 'r') as zip_ref:
    zip_ref.extractall('/content/price_data')

print("Extract selesai 🚀")
```

3. Cek letak folder yang terlah di exctract

4. Combaine beberapa file csv

```
import pandas as pd
import os

folder_path = "/content/price_data/price_data"

all_data = []

for file in os.listdir(folder_path):
    if file.endswith(".csv"):
        file_path = os.path.join(folder_path, file)

        df = pd.read_csv(file_path)

        # rename kolom biar konsisten
        df.columns = df.columns.str.lower().str.strip()
        df = df.rename(columns={"symbol": "ticker"})

        all_data.append(df)

final_df = pd.concat(all_data, ignore_index=True)

# format tanggal (penting)
final_df['date'] = pd.to_datetime(final_df['date'])

# simpan
final_df.to_csv("combined_stock_prices.csv", index=False)

print("DONE 🚀")
```
